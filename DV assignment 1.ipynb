```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('canteen_shop_data.csv')
df
```

```python
sns.scatterplot(data=df, x='Employee ID', y='Payment Method')
plt.title('Employee ID vs Payment Method')
plt.show()
```

```python
sns.histplot(data=df, x='Weather', bins=20, kde=True)
plt.title('Weather	')
plt.show()
```

```python
sns.boxplot(data=df, x='Customer Satisfaction', y='Special Offers')
plt.title('customer satisfaction by special offer')
plt.show()
```

```python
sns.violinplot(data=df, x='Customer Satisfaction', y='Special Offers')
plt.title('customer satisfaction')
plt.show()
```

```python
sns.barplot(data=df, x='Customer ID', y='Price', estimator=np.mean)
plt.title('customer id and price')
plt.show()
```

```python
sns.pairplot(df[['Total', 'Employee ID','Customer Satisfaction','Quantity' ]])
plt.title('Pairplot')
plt.show()
```
